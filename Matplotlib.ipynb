## Matplotlib

Matplotlib é uma biblioteca para criação de gráficos e visualização de dados em Python. Durante a exploração dos dados e na avaliação de resultados é muito útil usar gráficos para entender melhor o que está acontecendo. Visualização de dados é por si só uma área grande da ciência de dados, então vamos ver apenas uma introdução com os conceitos mais importantes.

```python
import numpy as np
import pandas as pd
# Só iremos usar o modulo pyplot e iremos chamá-lo de plt
import matplotlib.pyplot as plt
```

Vamos começar com um exemplo simples de um gráfico de linha e a partir dele explorar as funcionalidades do *matplotlib*

```python
# Vamos definir alguns dados ficticios usando listas
empresa_a = [1000, 1500, 1200, 3000, 2100, 2000]
empresa_b = [750, 2300, 1000, 500, 1200, 1500]
```

Para fazer um gráfico de barras usamos a função `plt.plot()`

```python
# Criar o gráfico
# posso passar a marcação dos indices do eixo 'x', como se fosse uma list, como no exemplo a baixo: [10,20,30] e etc
plt.plot([10, 20 ,30, 40, 50, 60],empresa_a)
```

```python
# Podemos aumentar o tamanho da imagem
# largura a esquerda e altura a direita
plt.figure(figsize=(10,6))

# Criar o gráfico
# caso queira evitar essa linha de retorno a baixo, do matplolib... coloque um ; no fim da chamada
plt.plot(empresa_a, marker='o', linestyle='dashed', color='deepskyblue' );
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 5))
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun']

# Criar o gráfico alterando o estilo
plt.plot(meses, empresa_a, marker='o', linestyle='dashed', linewidth=2)

# Definir um titulo para o gráfico
plt.title('Lucos da Empresa A')

# Definir a legenda de cada eixo
plt.xlabel('Mes')
plt.ylabel('Lucro');
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10,6))

# Criar os gráfico com estilo e legenda
plt.plot(meses, empresa_a, '--o', label='Empresa A', color='green')
plt.plot(meses, empresa_b, '--o', label='Empresa B', color='blue')

# Definir um titulo para o gráfico
plt.title('Lucros da Empresa A e B')

# Definir a legenda de cada eixo
plt.xlabel('Mês')
plt.ylabel('Lucro')

# Indica que queremos colocar a legenda
plt.legend(loc=4);
```

### Visualizações univariadas

Para essa visualizações iremos utilizar o conjunto de dados [Palmer Penguins](https://github.com/allisonhorst/palmerpenguins) que contém informações sobre pinguins de três espécies diferentes.

```python
# Vamos carregar o conjunto de dados que vamos usar
df = pd.read_csv('data/penguins_size.csv')
print(df.shape)
df.tail()
```

#### Gráfico de barra
Primeiro vamos fazer um exemplo simples com dados inventados

```python
# Altura de cada barra
altura = [12,2,8,5,7]

# Posição de cada barra
posicao = [5,8,7,9,6]

# Plotando o gráfico
plt.bar(posicao, altura);
```

Agora vamos visualizar o número de pinguins de cada espécie utilizando um gráfico de barras

```python
# .value_counts() conta a frequencia de cada valor
contagem_especies = df['species'].value_counts()
contagem_especies
```

```python
especies = list(contagem_especies.index)
valor_especies = contagem_especies.values

print(especies)
print(valor_especie)
```

```python
# Plotando as barras
plt.bar(np.arange(len(especies)), valor_especies)

# Mudando os valores no eixo x
plt.xticks(np.arange(len(especies)), especies);
```

#### Histograma

Um histograma mostra a distruibuição de uma variável, isso é, a frequencia em que cada uma dos seus valores ocorre. Usando `plt.hist` vamos gerar um histograma para o comprimento dos bicos dos pinguins (`culmen_length_mm`).

```python
# Basta passarmos a variavel de interesse
plt.hist(df['culmen_depth_mm'])
```

```python
# Podemos mudar o tamanho dos intervalos que são considerados para contagem
plt.hist(df['culmen_depth_mm'], bins=50)
```

#### Scatterplot

Essa é uma maneira de visualizar a relação entre duas variáveis. Vamos usar a função `plt.scatter` para ver a relação entre o comprimento das asas (`flipper_length_mm`) e a massa corporal (`body_mass_g`)

```python
# Ploatando os dados
plt.scatter(df['body_mass_g'], df['flipper_length_mm'])

# Nomeando os eixos
plt.xlabel('Massa Corporal')
plt.ylabel('Tamanho das Asas')
```
